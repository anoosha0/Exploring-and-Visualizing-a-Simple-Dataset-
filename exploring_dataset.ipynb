```python
# Exploring and Visualizing a Simple Dataset 
# Objective:To load, inspect, and visualize the Iris dataset to understand its structure, feature distributions,
# and inter-feature relationships using data exploration and basic plotting techniques.
```

```python
#Dataset Loading and Preprocessing
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
iris = sns.load_dataset('iris')

# Display basic info
print("Shape of dataset:", iris.shape)
print("\nColumn names:", iris.columns.tolist())
print("\nFirst 5 rows of data:")
print(iris.head())

# Check info and summary statistics
print("\nDataset Info:")
print(iris.info())
print("\nDescriptive Statistics:")
print(iris.describe())
```

```python
#Scatter Plot (Sepal Length vs Sepal Width)
plt.figure(figsize=(6, 4))
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title("Sepal Length vs Width by Species")
plt.show()
```

```python
#Histograms for All Features
iris.hist(figsize=(10, 6), bins=15, edgecolor='black', color='lightgreen')
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

```python
#Box Plots to Detect Outliers
plt.figure(figsize=(12, 6))
for i, col in enumerate(iris.columns[:-1], 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=iris, x='species', y=col)
    plt.title(f'Box Plot of {col}')
plt.tight_layout()
plt.show()
```
